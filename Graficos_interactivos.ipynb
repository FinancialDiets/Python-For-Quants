```python
## Veamos stacked bars
import plotly.graph_objects as go

animals=['giraffes', 'orangutans', 'monkeys']



## definimos la figura
fig = go.Figure(data=[
    go.Bar(name='SF Zoo', x=animals, y=[20, 14, 23]),
    go.Bar(name='LA Zoo', x=animals, y=[12, 18, 29])
])

# Change the bar mode
fig.update_layout(barmode='stack')
fig.show()
```

```python
import plotly.graph_objects as go

x = ['Product A', 'Product B', 'Product C']
y = [20, 14, 23]

# Agregamos más texto a las ventanas emergentes
fig = go.Figure(data=[go.Bar(x=x, y=y,
            hovertext=['27% market share', '24% market share', '19% market share'])])

# Personalizamos el gráfico
fig.update_traces(marker_color='rgb(158,202,225)', marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6)

fig.update_layout(title_text='January 2013 Sales Report')
fig.show()
```

```python
## Agregamos texto a las barras
import plotly.graph_objects as go

x = ['Product A', 'Product B', 'Product C']
y = [20, 14, 23]

#  textposition='auto' para la dirección del texto
fig = go.Figure(data=[go.Bar(
            x=x, y=y,
            text=y,
            textposition='auto',
        )])

fig.show()
```

```python
## Rotamos un poco los ejes
import plotly.graph_objects as go

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

## Definimos la figura
fig = go.Figure()
## Agregamos la primera barra
fig.add_trace(go.Bar(
    x=months,
    y=[20, 14, 25, 16, 18, 22, 19, 15, 12, 16, 14, 17],
    name='Primary Product',
    marker_color='indianred'
))
## Agregamos la segunda barra
fig.add_trace(go.Bar(
    x=months,
    y=[19, 14, 22, 14, 16, 19, 15, 14, 10, 12, 12, 16],
    name='Secondary Product',
    marker_color='lightsalmon'
))

# modificamos la rotación del eje con xaxis_tickangle
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

## 2. Gráfico de pie y donut

```python
import plotly.express as px
## extraemos la data
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
## a paises con población baja las renombramos
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries'
df.head()
```

```python
## definimos el gráfico y sus parámetros
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
```

```python
fig = px.pie(df, values='tip', names='day', color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

```python
## Personalizando el gráfico
import plotly.express as px
df = px.data.gapminder().query("year == 2007").query("continent == 'Americas'")

fig = px.pie(df, values='pop', names='country',
             title='Population of American continent',
             hover_data=['lifeExp'], labels={'lifeExp':'life expectancy'})
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

Ahora usemos go:

```python
import plotly.graph_objects as go

labels = ['Oxygen','Hydrogen','Carbon_Dioxide','Nitrogen']
values = [4500, 2500, 1053, 500]

fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.show()
```

```python
import plotly.graph_objects as go
## colores
colors = ['gold', 'mediumturquoise', 'darkorange', 'lightgreen']

## definimos la figura
fig = go.Figure(data=[go.Pie(labels=['Oxygen','Hydrogen','Carbon_Dioxide','Nitrogen'],
                             values=[4500,2500,1053,500])])
## diseño del grafico
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=20,
                  marker=dict(colors=colors, line=dict(color='#000000', width=2)))

fig.show()
```

```python
## Donut
import plotly.graph_objects as go

labels = ['Oxygen','Hydrogen','Carbon_Dioxide','Nitrogen']
values = [4500, 2500, 1053, 500]

# usamos "hole" para crear el circulo del medio
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=0.3)])
fig.show()
```

```python
## subplots

## importamos paquetes
import plotly.graph_objects as go
from plotly.subplots import make_subplots

## labels
labels = ["US", "China", "European Union", "Russian Federation", "Brazil", "India",
          "Rest of World"]

# Creamos los subplots: usamos 'domain' type para el Pie subplot
fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]])
## agregamos la primera dona
fig.add_trace(go.Pie(labels=labels, values=[16, 15, 12, 6, 5, 4, 42], name="GHG Emissions"),
              1, 1)
## agregamos la segunda dona
fig.add_trace(go.Pie(labels=labels, values=[27, 11, 25, 8, 1, 3, 25], name="CO2 Emissions"),
              1, 2)

# ponemos el 'hole'
fig.update_traces(hole=.4, hoverinfo="label+percent+name")

fig.update_layout(
    title_text="Global Emissions 1990-2011",
    # Agregamos anotaciones en el centro de las donas
    annotations=[dict(text='GHG', x=0.18, y=0.5, font_size=20, showarrow=False),
                 dict(text='CO2', x=0.82, y=0.5, font_size=20, showarrow=False)])
fig.show()
```

```python
## Otro ejemplo

import plotly.graph_objects as go
from plotly.subplots import make_subplots

## labels
labels = ["Asia", "Europe", "Africa", "Americas", "Oceania"]

## Definimos el subplot
fig = make_subplots(1, 2, specs=[[{'type':'domain'}, {'type':'domain'}]],
                    subplot_titles=['1980', '2007'])
## primera dona
fig.add_trace(go.Pie(labels=labels, values=[4, 7, 1, 7, 0.5], scalegroup='one',
                     name="World GDP 1980"), 1, 1)
## segunda dona
fig.add_trace(go.Pie(labels=labels, values=[21, 15, 3, 19, 1], scalegroup='one',
                     name="World GDP 2007"), 1, 2)

fig.update_layout(title_text='World GDP')
fig.show()
```

```python
import plotly.graph_objects as go

# Creamos dandom data con numpy
import numpy as np
np.random.seed(1)

N = 100 ##total de numeros
random_x = np.linspace(0, 1, N) ## eje x
random_y0 = np.random.randn(N) + 5
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N) - 5
```

```python
# creamos la figura
fig = go.Figure()
### agreagamos la primera linea
fig.add_trace(go.Scatter(x=random_x, y=random_y0,
                    mode='lines',
                    name='lines'))
### agreagamos la segunda linea
fig.add_trace(go.Scatter(x=random_x, y=random_y1,
                    mode='lines+markers',
                    name='lines+markers'))
### agreagamos la tercera linea
fig.add_trace(go.Scatter(x=random_x, y=random_y2,
                    mode='markers', name='markers'))

fig.show()
```
